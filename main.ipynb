## Memory — multi-turn conversations with ChatSession 🧠

`run_agent()` starts fresh every call — great for one-off tasks, but Claude
forgets context between calls.

`ChatSession` keeps a `history` list that grows with every exchange.
Before each call, it prepends the full history so Claude sees the whole thread.

**Rule of thumb:**
- Use `run_agent()` for discrete tasks → "Screen PRJ-001", "Add this participant"
- Use `ChatSession` for follow-up conversations → "Now shortlist Priya too", "Draft emails for both"

````python
from agent import run_agent, ChatSession
````

````python
# Replay the exact conversation that failed — this time with memory
# Notice: we create the session once and reuse it across all three messages

session = ChatSession()

print("=== Message 1 ===")
print(session.chat("I need 2 more participants for the Checkout Flow project. Who in the panel haven't we worked with yet and would be a good fit?"))
print()

print("=== Message 2 ===")
print(session.chat("Could you shortlist Priya and draft outreach emails for both her and Maya please?"))
print()

print("=== Message 3 ===")
print(session.chat("It's for Checkout Flow Redesign. Thank you!"))
````

````python
# You can also peek at what's in the history after a conversation
# Each item is a message Claude either sent or received
print(f"Messages in session history: {len(session.history)}")
for msg in session.history:
    role = msg["role"]
    # Content can be a string or a list of blocks (for tool calls)
    if isinstance(msg["content"], str):
        preview = msg["content"][:80]
    else:
        preview = f"[{len(msg['content'])} block(s) — tool calls/results]"
    print(f"  {role}: {preview}")
````

````python
# Reset when you want to start a new topic
session.reset()

# Now it's a clean slate — same session object, empty history
print(session.chat("Show me everyone in the panel."))
````

## Phase 7 — CSV import 📥

Bulk-add participants from a spreadsheet instead of one at a time.

**New concept: the `csv` module**
Python has a built-in `csv` module — no installation needed.
`csv.DictReader` reads each row as a dictionary, so `row["email"]` just works.

**The tricky bit: `preferred_methods`**
In JSON it's a list: `["interview", "usability test"]`
In CSV it's a string in one cell: `"interview, usability test"`
The tool splits it back into a list automatically.

**Try it two ways:**

## Organisation schema 🏢

**The new concept: a third JSON file joining the relational web**
```
participants.json  ←→  organisations.json  ←→  (projects.json later)
```
Participants have an `organisation_id` field that links to an org record.
When you call `get_organisation`, it cross-references both files to show
who from that company is in your panel — same VLOOKUP pattern, third file.

**Backward compatible:** participants without `organisation_id` still work fine.
The string `organisation` field stays as a display name cache.
